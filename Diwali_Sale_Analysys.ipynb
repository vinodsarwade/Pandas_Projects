<h1>Exploratory Data Analysis

<h2>Gender

```python
df.columns
```

```python
var = sns.countplot(x = "Gender", data = df)  #countplot at Gender column 

for i in var.containers:  #this loop is for to print values on graph
    var.bar_label(i)
```

```python
df.groupby(["Gender"], as_index = False) ["Amount"].sum() .sort_values( by = "Amount", ascending = False)
```

```python
sales = df.groupby(["Gender"], as_index = False) ["Amount"].sum() .sort_values( by = "Amount", ascending = False)
sns. barplot(x = "Gender", y = "Amount" , data = sales)
```

<h2>Age

```python
df.columns
```

```python
ax = sns.countplot(x = "Age Group", hue="Gender", data = df) # without hue also it can work but it can not show male and female. it shows overall age group bar.
```

```python
ax = sns.countplot(x = "Age Group", hue="Gender", data = df)

for i in ax.containers:
    ax.bar_label(i)
```

```python
sales_age = df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False)
sns.barplot(x = "Age Group", y = "Amount", data=sales_age)
```

<h8>from above graph we can see that most of the buyers are of age grop between 26-35 years. female shopers as > male.

<h2>State

```python
df.columns
```

```python
#total numbers of orders from each state
sales_state = df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by = "Orders", ascending=False)
sns.set(rc={"figure.figsize":(15, 5)})  # sometimes your graph names will overlaping to avoid that we can give size like this.
sns.barplot(x = "State", y = "Orders", data= sales_state)
```

```python
sales_state = df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False,).head(10)
sns.set(rc={"figure.figsize":(15, 5)})  # sometimes your graph names will overlaping to avoid that we can give size like this.
sns.barplot(x = "State", y = "Amount", data= sales_state)
```

<h8>*from abobe graph we can see that most of the orders from UP, maharashtra, and karnataka respectively by sales/amount.

<h2>Marital status

```python
df.columns
```

```python
ax = sns.countplot(x = "Marital_Status", data=df , hue="Gender")
sns.set(rc={"figure.figsize":(7, 5)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
status = df.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False)
sns.barplot(x = "Marital_Status", y = "Amount", data=status, hue="Gender")
```

From above Graph we can see that maritals people are more purchased than unmarried people. in that  marital female have more purchased than male. 

<h2>Occupation

```python
df.columns
```

```python
ax = sns.countplot(x="Occupation", data=df)
sns.set(rc={'figure.figsize':(20,5)})
for i in ax.containers:
    ax.bar_label(i)
```

```python
occ = df.groupby("Occupation", as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False)
sns.barplot(x = "Occupation", y = "Amount", data=occ)
```

As we can see here most of the amount has been spend from IT sector , Healthcare respectively.

<h2>Product Category

```python
ax = sns.countplot(x ="Product_Category", data = df)

for i in ax.containers:
    ax.bar_label(i)
```

```python
product = df.groupby(["Product_Category"],as_index=False)["Amount"].sum().sort_values(by = "Amount", ascending=False)
sns.barplot(x = "Product_Category", y = "Amount", data=product)
```

from above graph we can see that most of the sold products from Food, clothing and electronics Category.

```python
#find top selling product
selling = df.groupby(["Product_ID"],as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
sns.barplot(x= "Product_ID", y = "Amount", data=selling)
```

```python
df.groupby("Product_ID")["Amount"].sum().nlargest(10).sort_values(ascending=False).plot(kind = "bar")
```
