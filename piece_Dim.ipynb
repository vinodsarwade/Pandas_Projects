<h2>Line and scatter plot

```python
plt.plot(df['Length'])
mean_l = df["Length"].mean()
sd_l = df["Length"].std()
plt.axhline(y=mean_l, color = 'red', linestyle = '--')
plt.axhline(y=(mean_l+3*sd_l), color = 'red', linestyle = '--')
plt.axhline(y=(mean_l-3*sd_l), color = 'red', linestyle = '--')
plt.show()
```

```python
plt.scatter(x = df["Length"], y = df["Width"],marker='v')
plt.show()
```

<h1>Seaborn Displot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv("C:\\Users\\VSARWADE\\Desktop\\Pandas_project\\Piece_Dim.csv")
df.head()
```

```python
sns.histplot(data=df,x = 'Length',bins=10)
plt.title('Histogram by seaborn')
plt.show()
```

```python
sns.displot(data=df,x = 'Length',bins=10)
plt.title('Histogram by seaborn')
plt.show()
```

```python
sns.displot(data=df,x = 'Length',bins=10,hue='Operator')
plt.show()
```

```python
sns.displot(data=df,x = 'Length',bins=10,col ='Operator')
plt.show()
```

```python
sns.displot(data=df, x='Length',kind='kde',rug=True)
```

```python
sns.displot(data=df, x='Length',kind='kde',rug=True, hue='Operator')
```

```python
sns.displot(data=df, x='Length',kind='hist',kde= True,rug=True)
```
