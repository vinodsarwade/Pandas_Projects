<h2>Convert Datatype of rate column string -> float

```python

def handleData(value):
    value= str(value).split("/")
    value = value[0]
    return float(value)

dataframe["rate"] = dataframe["rate"].apply(handleData)
dataframe
```

<h2> Q1:Type of resturant

```python
dataframe
```

```python
var = sns.countplot(x=dataframe["listed_in(type)"])
plt.xlabel("type of resturant")

for i in var.containers:
    var.bar_label(i)
```

<h4>conclusion: Mejority of resturant falls under dining category

<h4>Q2: How many votes has each type of restuarant received from cunstomer

```python
dataframe.head()
```

```python
group_data = dataframe.groupby("listed_in(type)")["votes"].sum()  #group both column
# result = pd.DataFrame({"Votes by cust":group_data})    # you can make new DF using group_data  with column name as "votes by cust "
plt.plot(group_data, color ="green", marker="o")
plt.xlabel("Type of Restaurant", color= "red", size = "20")
plt.ylabel("votes by customer", color = "red", size = "20")
group_data

#'''to show count on the plot'''
for i, vote_count in enumerate(group_data):
    plt.text(group_data.index[i],vote_count,str(vote_count))
plt.show()
```

<h5>Conclusion: dinning resturants has received maximum votes

<h2>Q5: which mode received maximum rating online/offline

```python
dataframe.head()
```

```python
# var = plt.figure(figsize=(6,6))                          #using boxplot
# sns.boxplot(x="online_order",y="rate",data=dataframe)

max_rating = dataframe.groupby('online_order')['rate'].mean()  #using groupby
print(max_rating)
sns.barplot(max_rating)
# for i, rate in enumerate(max_rating):
#     plt.text(max_rating.index[i],rate,str(rate))

for i, rate in enumerate(max_rating.values):
    plt.text(i,rate, str(rate), ha='center', va= 'bottom')
```
